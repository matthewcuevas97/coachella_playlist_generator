```python
genre_to_id = {}
id_to_genre = {}
id_to_name = {}
for index, row in lineup_artists_pandas.iterrows():
    genres, spotify_id, name = row['genres'], row['id'], row['name']
    id_to_genre[spotify_id] = genres 
    id_to_name[spotify_id] = name
    for genre in genres:
        if genre in genre_to_id:
            genre_to_id[genre] += [spotify_id]
        else:
            genre_to_id[genre] = [spotify_id]
```

```python
coachella_genres = reduce(lambda x,y: x+y,lineup_artists_pandas['genres'])
genres = [id_to_genre[spotify_id] for spotify_id in ids if spotify_id in id_to_genre.keys()] + list(my_top_artists['genres'])
genres = reduce(lambda x,y: x+y,genres)
genres = [g for g in genres if g in coachella_genres]
genre_match_num = len(genres)
genre_count = Counter(genres)
matched_genre_proportion = np.array([[g , genre_count[g] * 1.0 / genre_match_num] for g in genre_count])
matched_genre_proportion
```
